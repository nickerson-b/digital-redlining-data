Try and get CRSF token tied to this dwsid

```python
# this gets crsf tied with dwsid
cookies = {
    'dwsid': dwsid
}

headers = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36',
}

response = requests.get('https://www.quantumfiber.com/shop/smartNidSpeeds', cookies=cookies, headers=headers)
print([resp.url for resp in response.history])
# qf-tile__price-amount
# cq_params.products
```

```python
# extract csrf
g = response.iter_content(decode_unicode=True)
line = []
for item in g:
    if item == '\n':
        line = ''.join(line)
        if ('csrf' in line):
            break
        line = []
    else:
        line.append(item)


tree = etree.HTML(line)
csrf = tree.find(".//input").get('value')
print(csrf)
```
